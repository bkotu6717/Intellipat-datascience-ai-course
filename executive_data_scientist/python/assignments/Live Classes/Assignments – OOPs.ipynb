```python
# Q1. Write a Python program to demonstrate multiple inheritance.

# 1. Employee class has 3 data members EmployeeID, Gender (String), Salary and
# PerformanceRating(Out of 5) of type int. It has a get() function to get these details from
# the user.
# 2. JoiningDetail class has a data member DateOfJoining of type Date and a function
# getDoJ to get the Date of joining of employees.

# 3. Information Class uses the marks from Employee class and the DateOfJoining date
# from the JoiningDetail class to calculate the top 3 Employees based on their Ratings
# and then Display, using readData, all the details on these employees in Ascending
# order of their Date Of Joining.

import datetime

def getKey(obj):
    return obj.performance_rating
    
def sortKey(obj):
    return obj.DateOfJoining

class Employee:
    def __init__(self, employee_id, gender, performance_rating):
        self.employee_id = employee_id
        self.gender = gender
        self.performance_rating = performance_rating
    
    def get(self):
        return self.__dict__
 

class JoiningDetail(Employee):
    def __init__(self, id, gender , performance_rating, doj):
        Employee.__init__(self, id, gender , performance_rating)
        
        self.DateOfJoining = datetime.datetime.strptime(doj, "%d%m%Y").date()
    def getDoJ(self):
        return self.DateOfJoining

class Information(JoiningDetail):
    def __init__(self):
        pass

    def readData(self, emps):
        print("Sort by performance_rating")
        emps.sort(key= getKey, reverse=True)
        for x in emps[0:3]:
            print(x.employee_id, x.gender, x.performance_rating, x.DateOfJoining)
        print('Sort by DateofJoinging')
        emps = emps[0:3]
        emps.sort(key= sortKey, reverse=False)
        for x in emps[0:3]:
            print(x.employee_id, x.gender, x.performance_rating, x.DateOfJoining)
        

doj = JoiningDetail(1, 'male', 6, '24112023')
doj2 = JoiningDetail(2, 'female', 4, '23112023')
doj3 = JoiningDetail(3, 'female', 1, '22112023')
doj4 = JoiningDetail(4, 'male', 7, '22112020')

emps = [doj, doj2, doj3, doj4]

inf = Information()
inf.readData(emps)

```

```python
# Q.2 Write a Python program to demonstrate Polymorphism.

# 1. Class Vehicle with a parameterized function Fare, that takes input value as fare and
# returns it to calling Objects.
# 2. Create five separate variables Bus, Car, Train, Truck and Ship that call the Fare
# function.
# 3. Use a third variable TotalFare to store the sum of fare for each Vehicle Type.
# 4. Print the TotalFare.

import random

class Vehicle:
    def fare(self, amount):
        self.fare_amount = amount
        return self.fare_amount

class Bus(Vehicle):
    pass
class Car(Vehicle):
    pass
class Train(Vehicle):
    pass
class Truck(Vehicle):
    pass
class Ship(Vehicle):
    pass

def totalFare(objs):
    total_amount = 0
    for x in objs:
        total_amount += x.fare(random.randint(100,900))
    return total_amount

bus = Bus()
car = Car()
train = Train()
truck = Truck()
ship = Ship()

amount = totalFare([bus, car, train, truck, ship])

print('Total fare amount: ', amount)
```

```python
# Q3. Consider an ongoing test cricket series. Following are the names of the players and their
# scores in the test1 and 2.
# Test Match 1 :
# Dhoni : 56 , Balaji : 94
# Test Match 2 :
# Balaji : 80 , Dravid : 105
# Calculate the highest number of runs scored by an individual cricketer in both of the matches.

# Create a python function Max_Score (M) that reads a dictionary M that recognizes the player
# with the highest total score. This function will return ( Top player , Total Score ) . You can
# consider the Top player as String who is the highest scorer and Top score as Integer .

# Input : Max_Score({‘test1’:{‘Dhoni’:56, ‘Balaji : 85}, ‘test2’:{‘Dhoni’ 87, ‘Balaji’’:200}})
# Output : (‘Balaji ‘ , 200)

import pandas as pd

def format_data(matches):
    players_data = []
    for x,y in matches.items():
        data = {}
        players = list(y.keys())
        scores = list(y.values())
        for a in range(0, len(players)):
            data['match'] = x
            data['player'] = players[a]
            data['score'] = scores[a]
            players_data.append(data)
            data = {}
    return players_data

def MaxScoreByIndividual(matches, player):
    data = format_data(matches)
    data = sorted(data, key=lambda item: item['score'], reverse=True)
    match_data = [d for d in data if d['player'] == player]
    return (match_data[0]['match'], match_data[0]['player'], match_data[0]['score'])
    
def MaxScoreByMatch(matches, target_match):
    data = format_data(matches)
    data = sorted(data, key=lambda item: item['score'], reverse=True)
    match_data = [d for d in data if d['match'] == target_match]
    return (match_data[0]['player'], match_data[0]['score'])

matches = {
    'test1':{'Dhoni':56, 'Balaji' : 85}, 
    'test2':{'Dhoni': 87, 'Balaji':200}
}
print('Max score in test1: ', MaxScoreByMatch(matches, 'test1'))
print('Max score in test2: ',MaxScoreByMatch(matches, 'test2'))

print('Max score by Dhoni: ',MaxScoreByIndividual(matches, 'Dhoni'))
print('Max score by Balaji: ',MaxScoreByIndividual(matches, 'Balaji'))


# df = pd.DataFrame(matches)
# print(df)
# print('*'*100)
# data_set = MaxScore(matches)
# print(cricket_series)
```

```python
# Q4. Create a simple Card game in which there are 8 cards which are randomly chosen from a
# deck. The first card is shown face up. The game asks the player to predict whether the next card
# in the selection will have a higher or lower value than the currently showing card.

# For example, say the card that’s shown is a 3. The player chooses “higher,” and the next card is
# shown. If that card has a higher value, the player is correct. In this example, if the player had
# chosen “lower,” they would have been incorrect. If the player guesses correctly, they get 20
# points. If they choose incorrectly, they lose 15 points. If the next card to be turned over has the
# same value as the previous card, the player is incorrect.

class CardGame:
    def __init__(self):
        self.card_numbers = random.sample(range(1, 100), 8)
        self.target_card = self.card_numbers[0]
        self.score = 0
    
    def calculate_score(self, next_card):
        print('target card: ', self.target_card)
        print('next card: ', next_card)

        if next_card <= self.target_card:
            self.score -= 15
        else:
            self.score += 20
        self.target_card = next_card
        print('updated score: ', self.score)

    def play_game(self):
        for x in self.card_numbers[1:]:
             self.calculate_score(x)
        print(self.score)

CardGame().play_game()
            
    
```

```python
# Q6. Show a basic implementation of abstraction in python using the abstract classes.
# 1. Create an abstract class in python.
# 2. Implement abstraction with the other classes and base class as abstract class.

from abc import ABC, abstractmethod

class Polygon(ABC):
 
    @abstractmethod
    def noofsides(self):
        pass

class Triangle(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 3 sides")
 
 
class Pentagon(Polygon):
 
    # overriding abstract method
    def noofsides(self):
        print("I have 5 sides")

R = Triangle()
R.noofsides()

R = Pentagon()
R.noofsides()
```
