```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix
import seaborn as sns
```

```python
data = pd.read_csv('./Files/iris.csv')
```

```python
data.head()
```

## Step 5: Analyze the shape of data

```python
data.shape
```

```python
data.isnull().sum()
```

## Step 6: Separate data into X and Y variables and split them into training and
## testing set with 70/30 ratio

```python
x, y = data.iloc[:, :-1], data.iloc[:, -1]
```

```python
x
```

```python
y
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
x_train
```

```python
x_test
```

```python
y_train
```

## Step 7: Instantiate and train a logistic regression model

```python
y_test
```

```python
clf = LogisticRegression()
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
y_pred
```

```python
confusion_matrix(y_test, y_pred)
```

```python
accuracy_score(y_pred, y_test)
```
