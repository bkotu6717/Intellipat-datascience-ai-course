```python
# Problem Statement:
# You work in XYZ Company as a Python developer. The company officials want
# you to build a Python program.
# Tasks To Be Performed:


# 1. Load cars data as dataframe using pandas and create a bar plot between
# number of cylinders and frequency of cars with that many number of
# cylinders

# ● Set xlabel as Number of cylinders
# ● Set ylabel as Frequency of cars
# ● Draw a bar plot

import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

df = pd.read_csv('./cars.csv')
dictionary = Counter(df['cyl'])
plt.bar(dictionary.keys(), dictionary.values())
plt.xlabel('Number of cylinders')
plt.ylabel('Frequency of cars')
plt.title('frequency of cars with that many number of cylinders')
plt.show()



```

```python

# 2. Write code to load data from cars and print 
# a bar graph of count of columns with null values.
df = df.drop(['S.No'], axis=1)
nan_data = dict(df.isna().sum())
plt.bar(nan_data.keys(), nan_data.values())
```

```python
# 3. Use the 'mpg' (Miles Per Gallon column) and draw a histogram:
# ● Set xlabel: Miles per gallon
plt.xlabel('Miles per gallon')
# ● Set ylabel: Frequency
plt.ylabel('Frequency')
# ● Set title as Miles Per Gallon Histogram
plt.title('Miles Per Gallon Histogram')
# ● Use mpg column to generate a histogram
plt.hist(df['mpg'], bins=10, alpha=0.5, histtype="bar", ec="black")
# plt.xticks(range(10,35))
```
