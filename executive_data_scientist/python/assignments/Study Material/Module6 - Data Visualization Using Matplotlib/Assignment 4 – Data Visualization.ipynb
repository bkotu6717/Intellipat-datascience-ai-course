```python
# Problem Statement:
# You work in XYZ Corporation as a Data Analyst. Your corporation has told you to
# visualize the mtcars.csv dataset with various plots.

# Tasks To Be Performed:
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('./mtcars.csv')
df.rename(columns = {'Unnamed: 1':'Model'}, inplace = True)
df['Model'] = df['Model'].fillna('Unknown Model')
df['cyl'] = df['cyl'].fillna(0)

# print(df['Model'])

# 1. Plot a pie chart for columns: ‘cyl’ and ‘model’ from the mtcars.csv data
# frame.
plt.pie(df['cyl'], labels=df['Model'], radius=3, autopct='%1.1f%%', shadow=True)
plt.show()
```
