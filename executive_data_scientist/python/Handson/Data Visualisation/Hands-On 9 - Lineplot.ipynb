```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

x = [random.randrange(0,5,1) for x in range(0,10)]
y = [random.randrange(0,5,1) for x in range(0,10)]

plt.figure(figsize=(5,3), dpi=300)
plt.plot(x,y, marker='D')

x2 = np.arange(1,5,0.5)
y2 = x2**2
plt.plot(x2, y2)

# x_coordintes = [x for x in range(0,10)]
# y_coordinates = [x*2 for x in range(0,10)]

# plt.plot(x_coordintes, y_coordintes, marker='v')

plt.title('Basic line chart')
plt.xlabel('X')
plt.ylabel('Y')
plt.legend('X and Y')
plt.savefig('linegraph.png')
plt.show()
```
