```python
# Calculate Mean median mode by functioin
import numpy as np
import statistics as stats

# median
data=[80,85,90,92,88]
median=np.median(data)
print(median)

data2=[80,85,88,89,92,95]
median2=np.median(data2)
print(median2)

# Mean
data=[80,85,90,92,88]
mean=np.mean(data)
print(mean)

# Mode
data=[98,85,86,85,98,81,80,80,89]  # in case of values having the same frequency it will take the first value
mode=stats.mode(data)
print(mode)

# Range
data=[10,20,30,40,50,67,54,78,98]
range = np.ptp(data)
print(range)

# variance
data=[80,85,90,92,88]
variance=np.var(data)
print(variance)

# standard deviation
data=[80,85,90,92,88]
std=np.std(data)
print(std)
```
