```python
# map -> traverse entire list -> new list
# [1, 2, 3, 4, 5] -> [1, 4, 9, 16, 25]

nums = [1, 2, 3, 4, 5]
# sq_nums = []

# for num in nums:
#     sq_nums.append(num * num)
    
# print(sq_nums)


def square(num):
    return num * num

# map(fn, list)
sq_nums = map(square, nums)
print(list(sq_nums)) # map object

# for num in sq_nums:
#     print(num)
```

```python
# factorial upto 10

def get_factorial(num):
    factorial = 1
    for i in range(1, num+1):
        factorial *= i
    return factorial

factorials = map(get_factorial, range(1, 11))
print(list(factorials))
```

```python
def even(num):
    return num % 2 == 0

# filter -> boolean
even_nums = filter(even, [1, 2, 3, 4, 5])
print(list(even_nums))
```

```python
# higher order function -> a function which takes another fn as a param or returns another fn


def square():
    print('caculating square area')
    
def rectangle():
    print('caculating rectangle area')
    
def trapezium():
    print('caculating trapezium area')
    
# hof -> wrapper
def area(shape):
    shape()


# the fn which is passed as a param(square, rectangle) -> callback fns(future)
# print(type(square))
area(square)
area(rectangle)
area(trapezium)
```

```python
A = pd.DataFrame({
    'color': ['red', 'green', 'yellow'],
    'stock': [1, 2, 3],
})

B = pd.DataFrame({
    'color': ['yellow', 'green', 'pink'],
    'size': ['small', 'medium', 'large'],
})

# # SELECT * FROM A LEFT JOIN B ON A.color = B.color
# pd.merge(A, B, how='left')

# # SELECT * FROM A RIGHT JOIN B ON A.color = B.color
# pd.merge(A, B, how='right')

# SELECT * FROM A OUTER JOIN B ON A.color = B.color
pd.merge(A, B, how='outer')
```

```python
A = pd.DataFrame({
    'color': ['red', 'green', 'yellow'],
    'stock': [1, 2, 3],
})

B = pd.DataFrame({
    'tshirt': ['yellow', 'green', 'pink'],
    'size': ['small', 'medium', 'large'],
})

# SELECT * FROM A JOIN B ON A.color = B.tshirt

df = pd.merge(A, B, left_on='color', right_on='tshirt')
df
```
