```python
x = [2, 4, 6, 8, 10]
y = [1, 2, 3, 4, 5]

# figure -> always at the top
plt.figure(figsize=(10, 5), dpi=500) # dpi => dots per inch

plt.plot(x, y, label='salary', color='#10e6be', linestyle='-', linewidth=4, marker='h', markersize=10,
        markerfacecolor='red', markeredgecolor='black')

# shortcut
# color|marker|line
# plt.plot(x, y, 'b^-.')

# labels
plt.xlabel('salary in lakhs', fontdict={'fontname': 'Times New Roman', 'fontsize': 20})
plt.ylabel('Experience in years', fontdict={'fontname': 'Arial', 'fontsize': 20})
plt.title('IT salary standards', fontdict={'fontname': 'Gill Sans', 'fontsize': 20})

# ticks
# plt.xticks([1, 4, 10, 30, 70, 40])
# plt.yticks([1, 2, 3, 4, 5])
# plt.xticks(range(2, 11, 2))
# plt.yticks(range(1, 6))

# second line
x2 = np.array([1, 2, 3, 4, 5])
plt.plot(x2, x2**2, label='n^2')

plt.legend() # always write legends at the bottom


# save the image
plt.savefig('salary_standards.jpeg', dpi=100)
plt.savefig('salary_standards.png', dpi=100)


plt.show()
```

```python
# barplot
name = 'India won the match with pakistan'
# {a: 2, j: 1, y: 1}

char_map = {}

for char in name:
    if(char in char_map):
        char_map[char] += 1
    else:
        char_map[char] = 1
        
x = list(char_map.keys())
y = list(char_map.values())

plt.bar(x, y)

plt.xlabel('character')
plt.ylabel('frequency')

plt.show()
```

```python
x = ['a', 'b', 'c']
y = [3, 5, 2]

# plt.bar(x, y, color='black')
barplot = plt.bar(x, y, color=['red', 'green', 'blue'])
barplot[0].set_hatch('/')
barplot[1].set_hatch('.')
barplot[2].set_hatch('*')

plt.show()
```
