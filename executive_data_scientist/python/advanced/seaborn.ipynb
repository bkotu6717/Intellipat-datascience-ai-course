```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# read the CSV
df = pd.read_csv('./mtcars.csv')
df
```

```python
# bar plot
sns.barplot(x = df['cyl'], y = df['hp'])
plt.xlabel('Cylinder')
plt.ylabel('Horse Power')
plt.title('cyl vs hp')
plt.show()
```

```python
sns.barplot(x = df['cyl'], y = df['hp'], color='red')
plt.title('cyl vs hp')
plt.show()
```

```python
sns.barplot(x = df['cyl'], y = df['wt'], color='red')
plt.title('cyl vs wt')
plt.show()
```

```python
sns.barplot(x = df['cyl'], y = df['mpg'], color='red')
plt.title('cyl vs mileage')
plt.show()
```

```python
# palette
# sns.barplot(x = df['cyl'], y = df['mpg'], palette='Set1')
# sns.barplot(x = df['cyl'], y = df['mpg'], palette='Set2')
# sns.barplot(x = df['cyl'], y = df['mpg'], palette='Paired')
# sns.barplot(x = df['cyl'], y = df['mpg'], palette='rocket')
# sns.barplot(x = df['cyl'], y = df['mpg'], palette='mako')
# sns.barplot(x = df['cyl'], y = df['mpg'], palette='flare')
plt.show()
```

```python
plt.figure(figsize=(10, 5), dpi=400)
sns.barplot(x='cyl', y='mpg', data=df, color='red')
plt.show()
```

```python
# count plot
sns.countplot(x='cyl', data=df)
plt.show()
```

```python
sns.countplot(y='cyl', data=df)
plt.show()
```

```python
# correlation -> heatmap
# correlation -> both are +ve or both are -ve then -> +ve correlation
# correlation -> one is +ve and other is -ve then -> -ve correlation
df = df.loc[:, 'mpg':]

plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), linewidth=1, annot=True, cbar=False, cmap='Blues', vmin=-0.5, vmax=0.5)


plt.show()
```

```python
# distribution -> scatterplot
sns.scatterplot(x='disp', y='hp', data=df)
plt.show()
```

```python
sns.scatterplot(x='cyl', y='hp', data=df)
plt.show()
```

```python
sns.scatterplot(x='wt', y='hp', data=df)
plt.show()
```

```python
sns.scatterplot(x='carb', y='hp', data=df)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.scatterplot(x='mpg', y='hp', hue='Unnamed: 1', data=df)
plt.show()
```

```python
sns.scatterplot(x='mpg', y='hp', hue='wt', data=df)
plt.show()
```

```python
# pair plot
sns.pairplot(df)
plt.show()
```

```python
plt.figure(figsize=(20, 20))
sns.pairplot(df, hue='Unnamed: 1', palette='Set2')
plt.show()
```

```python
# boxplot -> to find the outlier
sns.boxplot(x='cyl', y='mpg', data=df)
plt.show()
```

```python
sns.boxplot(x='cyl', y='hp', data=df)
plt.show()
```
