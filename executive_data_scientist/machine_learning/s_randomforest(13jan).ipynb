```python
churn_data["TotalCharges"] = pd.to_numeric(churn_data["TotalCharges"], errors = 'coerce')

# " "  -> NAN
# ?    -> NAN
# -    -> NAN

# to_numeric function will change the datatype and errors = 'coerce will convert all the not convertable values into NAN
```

```python
churn_data = churn_data.dropna()
```

```python
col_list = []

for i in churn_data.columns:
  if((churn_data[i].dtype == 'object') & (i != "Churn")):
    col_list.append(i)

col_list
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()

for i in col_list:
  churn_data[i] = le.fit_transform(churn_data[i])
```

```python
# Model building part

# seperate independent and dependent columns
# train test split
# build our model --- checking for classification or regression models ----> fit_transform
# make predictions
# evaluating our model
# if a classification model :  accuracy score, confusion matrix, classification report
# if a regression model : r2 score, adjusted r2 score, mse, rmse, mae
```

```python
X = churn_data.iloc[:,:-1]
y = churn_data["Churn"]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier1 = RandomForestClassifier(n_estimators = 100) # number of decision trees = 100
classifier1.fit(X_train, y_train)
```

```python
predictions = classifier1.predict(X_test)
predictions
```

```python
y_test # these are actual values
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(predictions, y_test)
```

```python
n = [100,200,300,400,500,600,700,800,900,1000]

accuracy = []

for i in n:
  model = RandomForestClassifier(n_estimators = i)
  model.fit(X_train,y_train)
  pred = model.predict(X_test)
  a = accuracy_score(y_test,pred) *100
  accuracy.append(a)

accuracy
```

```python
d = {"Number of trees": n,
     "Accuracy" : accuracy}
df = pd.DataFrame(d)
df
```
